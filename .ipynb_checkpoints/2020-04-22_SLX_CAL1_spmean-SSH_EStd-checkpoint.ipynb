```python
def latexvarname(varna):
    if varna=='sosstsst':
        varname='SST'
        latexvarname=varname
    elif varna=='sossheig':
        varname='SSH'
        latexvarname=varname
    elif varna=='socurloverf':
        varname='curloverf'
        latexvarname="$\zeta/f$"
    else :
        varname=varna
        latexvarname=varna
    return varname,latexvarname
```
